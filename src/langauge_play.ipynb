```python
df = pd.DataFrame({"Hanzi": [], "Traditional": [], "Pinyin": [], "Pinyin 2": [], "Meaning": [], "Audio": []})
for line in lines:
    sections = line.split("\t")
    #row = {"Hanzi": sections[0], "Traditional": sections[1], "Pinyin": sections[7], "Pinyin 2": sections[8], "Meaning": sections[9], "Audio": sections[12]}
    row = pd.Series([sections[0], sections[1], sections[7], sections[8], sections[9], sections[12]], index=df.columns)
    df = pd.concat([df, row.to_frame().T], ignore_index=True)
```

```python
df = pd.read_csv(r"/home/david/Documents/Cyber-AI/data/mandarin_3000.csv")

for index, row in df.iterrows():

    if pd.isnull(df.loc[index, 'Traditional']):
        df.loc[index, "Traditional"] = df.loc[index, "Hanzi"]

    if not pd.isnull(df.loc[index, 'Pinyin 2']):
        string = df.loc[index, "Pinyin"] + ", " + df.loc[index, "Pinyin 2"]
        df.loc[index, "Pinyin"] = string
```

```python
final_df = pd.DataFrame({"Traditional": [], "Pinyin": [], "Meaning": [], "Audio": []})

for index, row in df.iterrows():
    characters = df.loc[index, "Traditional"].split()
    for character in characters:
        row = pd.Series([character, df.loc[index, "Pinyin"], df.loc[index, "Meaning"], df.loc[index, "Audio"]], index=final_df.columns)
        final_df = pd.concat([final_df, row.to_frame().T], ignore_index=True)
```

```python

final_df.to_csv(r"/home/david/Documents/Cyber-AI/data/Mandarin_3000_final.csv", index=False)
```

```python
final_df = pd.read_csv(r"/home/david/Documents/Cyber-AI/data/Mandarin_3000_final.csv")
```

```python
anki = pd.DataFrame({"1: Front": [], "2: Back": [], "tag": []})

for index, row in final_df.iterrows():

    if pd.isnull(final_df.loc[index, 'Pinyin']):
        continue

    back = final_df.loc[index, "Pinyin"]

    if not pd.isnull(final_df.loc[index, 'Meaning']):
        back = back + "\n" + final_df.loc[index, "Meaning"]
    if not pd.isnull(final_df.loc[index, 'Audio']):
        back = back + "\n" + final_df.loc[index, "Audio"]

    row = pd.Series([final_df.loc[index, "Traditional"], back, "tag"], index=anki.columns)
    anki = pd.concat([anki, row.to_frame().T], ignore_index=True)
```

```python
for index, row in anki.iterrows():
    anki.loc[index, "2: Back"] = anki.loc[index, "2: Back"].replace(",", "")
```
